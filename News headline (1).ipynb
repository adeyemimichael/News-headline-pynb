```python
#Import Necessary Libraries and Generate Random News Headlines
# For example, using an API to fetch real news headlines
import requests

def fetch_news_headlines():
    api_key = '31b16f72cece417f87201494fc846f00'  # Replace with your API key
    url = f'https://newsapi.org/v2/top-headlines?country=us&apiKey={api_key}'
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        headlines = [article['title'] for article in data['articles']]
        return headlines
    else:
        print("Failed to fetch headlines")
        return []
        
# Fetching real news headlines using an API
news_headlines = fetch_news_headlines()

```

```python
# Using TextBlob for sentiment analysis
from textblob import TextBlob

def analyze_sentiment(headlines):
    sentiment_scores = []
    for headline in headlines:
        analysis = TextBlob(headline)
        sentiment_scores.append(analysis.sentiment.polarity)
    return sentiment_scores

sentiment_scores = analyze_sentiment(news_headlines)
```

```python
This section imports TextBlob from the textblob library and defines a function analyze_sentiment() to analyze the sentiment of headlines using TextBlob sentiment analysis.
The function iterates through each headline, computes its sentiment polarity using TextBlob, and stores the scores in a list called sentiment_scores.
```

```python
# Visualizing sentiment scores
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(sentiment_scores, bins=20, color='skyblue')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.title('Distribution of Sentiment Scores in News Headlines')
plt.grid(axis='y')
plt.show()
```
